```python
# сгрупируем машины по возростным группам 
#(old = самые старые машины - 1990г, average = 1991 - 2010гг, young = 2011 - самая молодая машина)

min_age = data["year_produced"].min()
max_age = data["year_produced"].max()
conditions = [
    (data["year_produced"] >= min_age) & (data["year_produced"] <= 1990),
    (data["year_produced"] >= 1991) & (data["year_produced"] <= 2010),
    (data["year_produced"] >= 2011) & (data["year_produced"] <= max_age)]
choices = ['old', 'average', 'young']

data['AgeGroup'] = np.select(conditions, choicelist=choices)
data['AgeGroup'].value_counts()
```

```python
# количество машин в минском регионе по возрастным группам
cars_minsk = data.query("location_region == 'Минская обл.'").groupby('AgeGroup').agg({"year_produced" : "count"})
display(cars_minsk)
```
